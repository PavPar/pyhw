**- Найти выбросы. Удалить их.**

Если номер варианта четный - удалить всю строку данных из выборки.


```python
data = structData

low = .05
high = .95
quant_df = data.quantile([low, high])
data = data.apply(lambda x: x[(x>quant_df.loc[low,x.name]) & 
                                    (x < quant_df.loc[high,x.name])], axis=0).dropna()


print(structData.plot.line())
print(data.plot.line())
```

**- Найти новые средние и дисперсии, сравнить со старыми.**

```python
oldMean - data.mean()
```

Среднее значение увеличилось

```python
oldStd - data.std()
```

Дисперсия уменьшилась

**- Нормализовать данные по среднему и дисперсии.**

```python
df = data
normalized_df=(df-df.mean())/df.std())

normalized_df.plot()
```
