# Введение в Jupyter Notebook
Jupyter - среда для разработки, поддерживающая в т.ч. язык Python (но не только). Особенностью данной среды является ее разделение на серверную и клиентскую часть, где серверная часть работает с интерпретатором Python, а клиентская часть обеспечивает взаимодействие с пользователем.  
  
Давайте рассмотрим процесс работы в данной среде.  
  
Весь пользовательский ввод осуществляется с помощью ячеек - блоков для ввода текста, которые потом обрабатываются интерпретатором. Все ячейки связаны и работают на одном ядре ( - читайте, в одной консоли интерпретатора).  

Ячейки подразделяются на три основных типа, переключать который можно для каждой ячейки отдельно в шапке страницы. Формат raw мы рассматривать не будем, он - для html кода или сложных конструкций. Когда он вам понадобится - вы сами сможете найти по нему информацию.  
  
Эта и предыдущая ячейка - формата Markdown, специальный формат разметки текста. Текст в ней отображается по некоторым правилам форматирования (можно пару раз нажать на ячейку и посмотреть, как он выглядит в оригинале).

Цифры слева в каждой ячейке кода отображают, какой по счету эта ячейка была выполнена - специально для вашего удобства. Запланированные к выполнению или выполняющиеся ячейки помечены звездочкой (это не всегда правда, ячейка могла уже выполниться, а все еще со звездочкой, иногда Jupyter тупит).

Кратко вспомним остальные кнопки:
  * кнопка с дискетой - сохраняет данный лист
  * кнопка с плюсиком - добавляет ячейку
  * кнопка с ножницами - вырезает ячейку
  * следующие две кнопки - копирование ячейки и вставка
  * кнопки вверх и вниз перемещают ячейки на листе
  * кнопка Run - запускает выполнение кода
  * кнопка Stop (квадрат) - останавливает выполнение кода
  * далее кнопка с линией в виде почти завершенного круга - эта кнопка перезагружает весь лист, как будто вы закрыли и снова открыли консоль Python. Обычно используется, если что-то сломалось и вы хотите запустить снова и выполнять ячейки сами.
  * кнопка с перемоткой означает "перезапусти лист (ядро) и выполни все ячейки сверху до низу"
  * далее идет переключатель типов ячейки
  * и последняя кнопка вызывает командную панель, где есть всякие дополнительные команды
